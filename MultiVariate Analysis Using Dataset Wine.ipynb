```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DISPLAY_MAX_ROWS =1000
```

```python
data = pd.read_csv("wine.data", header=None)
data.columns = ["V"+str(i) for i in range(1, len(data.columns)+1)] 
X = data.loc[:, "V2":] 
y = data.V1     #Sample Number
data   #Actual Dataset
```

```python
data.loc[:, "V2":"V4"]   #Columns Used!
```

```python
pd.plotting.scatter_matrix(data.loc[:, "V2":"V4"])
plt.tight_layout()
plt.show()  
#Diagonal Shows the Histogram of two Variables
#Off-Diagonal shows the scatter points
```

```python
sns.lmplot("V4", "V5", data, hue="V1", fit_reg=False);

#CONCLUSION - Sample 3 has almost equal concentration of both V5 and V4, whereas sample 2 has lower concentration
#of V4 than compared to V5 and vice versa
```

```python
sns.lmplot("V9", "V14", data, hue="V1", fit_reg=False);

#CONCLUSION - Sample 1 has higher concentration of V14, whereas sample 2 AND 3 has higer concentration
#of V9 than compared to V5 and vice versa
```

```python
ax = data[["V2","V3","V4", "V6"]].plot()
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5));
#CONCLUSION - Shows that V6 has higher Mean and Variance than higer variable
```
